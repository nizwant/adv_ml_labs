```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
def generate_data_v1(n=1000, number_of_features=10, number_of_significant_features=3):
    X = np.random.randn(n, number_of_features)
    median_of_chi2 = chi2.median(number_of_significant_features)
    sum_of_squares = np.sum(
        X[:, :number_of_significant_features] ** 2, axis=1, keepdims=True
    )
    y = np.where(sum_of_squares > median_of_chi2, 1, 0)
    return X, y


def generate_data_v2(n=1000, number_of_features=10, number_of_significant_features=3):
    X = np.random.randn(n, number_of_features)
    sum_of_abs = np.sum(
        np.abs(X[:, :number_of_significant_features]), axis=1, keepdims=True
    )
    y = np.where(sum_of_abs > number_of_significant_features, 1, 0)
    return X, y
```

```python
# test generate_data_v1
X, y = generate_data_v1()
print(X, y)
print(X.shape, y.shape)
# distribution of y
# plt.hist(y)
# test generate_data_v2
X, y = generate_data_v2()
print(X, y)
print(X.shape, y.shape)
# distribution of y
plt.hist(y)
# distribution of y
```
