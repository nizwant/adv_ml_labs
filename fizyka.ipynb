```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
gl_dist = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
ol_dist = gl_dist

gl_time = [15, 16.32, 18.21, 19.34, 21.8, 23.5, 24.4, 25.75, 28.44, 29.34]
ol_time = [4.66, 5.12, 5.56, 5.94, 6.16, 6.93, 7.28, 7.69, 7.85, 8.59]
```

```python
plt.scatter(gl_time, gl_dist, label="Gliceryna", color="b")
plt.scatter(ol_time, ol_dist, label="Olej", color="orange")
plt.ylabel("odległość [cm]")
plt.xlabel("czas [s]")
plt.legend()
plt.title("Wizualizacja pomiarów czasu")
plt.grid()
```

```python
plt.errorbar(
    gl_time, gl_dist, xerr=0.5, yerr=0.2, label="Gliceryna", color="b", fmt="."
)
plt.errorbar(
    ol_time, ol_dist, xerr=0.5, yerr=0.2, label="Olej", color="orange", fmt="."
)

(m, b), cov = np.polyfit(gl_time, gl_dist, 1, cov=True)
plt.plot(gl_time, [m * x + b for x in gl_time], color="b", linewidth=2)
print("gliceryna")
print(m, b)
print(np.sqrt(cov))
(m, b), cov = np.polyfit(ol_time, ol_dist, 1, cov=True)
plt.plot(ol_time, [m * x + b for x in ol_time], color="orange", linewidth=2)
print("olej")
print(m, b)
print(np.sqrt(cov))

plt.ylabel("odległość [cm]")
plt.xlabel("czas [s]")
plt.legend()
plt.grid()
plt.title("Wizualizacja pomiarów czasu z regresją liniową i błędami pomiarowymi")
```
