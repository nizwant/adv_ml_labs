```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
copper_thickens = [2, 5, 7, 10, 12, 15, 17, 20]
al_thickens = [1, 2, 3, 5, 7, 10, 12, 15, 17, 20]
pb_thickens = [2, 5, 7, 10, 12, 15, 17, 20]
```

```python
copper_nums = [11251, 9943, 8593, 7138, 6549, 5560, 4975, 4277]
copper_log = [math.log(x) for x in copper_nums]
al_nums = [12470, 11925, 11835, 11363, 10972, 10489, 10157, 9457, 9307, 8794]
al_log = [math.log(x) for x in al_nums]
pb_nums = [10069, 7257, 5863, 4477, 3569, 2478, 1954, 1447]
pb_log = [math.log(x) for x in pb_nums]
```

```python
plt.scatter(copper_thickens, copper_nums, label='Miedź')
plt.scatter(al_thickens, al_nums, label="Aluminium")
plt.scatter(pb_thickens, pb_nums, label="Ołów")
plt.xlabel('x [mm]')
plt.ylabel('N')
plt.legend()
plt.title('Zależność zliczeń impulsów w zależności od grubości płytki')
plt.grid()
```

```python
copper_log_err = [1/math.sqrt(x) for x in copper_nums]
al_log_err = [1/math.sqrt(x) for x in al_nums]
pb_log_err = [1/math.sqrt(x) for x in pb_nums]
```

```python
plt.figure(figsize=(9, 7))
plt.errorbar(copper_thickens, copper_log, yerr=copper_log_err, label='Miedz', color='b', fmt='.')
plt.errorbar(al_thickens, al_log, yerr=al_log_err, label='Aluminium', color='orange', fmt='.')
plt.errorbar(pb_thickens, pb_log, yerr=pb_log_err, label='Ołów', color='g', fmt='.')
(m, b), cov = np.polyfit(copper_thickens, copper_log, 1, cov=True)
plt.plot(copper_thickens, [m * x + b for x in copper_thickens], color="b")
print("miedz")
print(m, b)
print(np.sqrt(cov))
(m, b), cov = np.polyfit(al_thickens, al_log, 1, cov=True)
plt.plot(al_thickens, [m * x + b for x in al_thickens], color="orange")
print("al")
print(m, b)
print(np.sqrt(cov))

(m, b), cov = np.polyfit(pb_thickens, pb_log, 1, cov=True)
plt.plot(pb_thickens, [m * x + b for x in pb_thickens], color="g")
print("olow")
print(m, b)
print(np.sqrt(cov))
plt.xlabel('x [mm]')
plt.ylabel('ln(N)')
plt.legend()
plt.grid()
plt.title('Logarytm zliczeń impulsów w zależności od grubości absorbentu dla różnych materiałów')
```
